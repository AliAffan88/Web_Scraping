```python
import pandas as pd
from tkinter import *
from tkinter import messagebox
from bs4 import BeautifulSoup
import urllib.request
from tkinter.scrolledtext import ScrolledText
import requests as rq

def scrape_data():                 # making function for working
    url = url_entry.get()      # taking the url from tkinter entry box 

    try:                                               # handling error exception
        response = rq.get(url)                         # getting the url  
        html_content = response.text                   # 

        if tag_var.get() == 0:
            tag_name = 'a'
        else:
            tag_name = 'p'

        soup = BeautifulSoup(html_content, 'html.parser')
        tags = soup.find_all(tag_name)

        data = []
        for tag in tags:
            text = tag.text
            href = tag.get('href')
            tag_class = tag.get('class')
            data.append([text, href, tag_class])

        df = pd.DataFrame(data, columns=['Text', 'Links', 'Class'])
        df_window = Toplevel()
        df_window.title('Scraped Data')
        df_window.geometry('800x600')

        text_widget = ScrolledText(df_window, width=100, height=30)
        text_widget.pack(fill=BOTH, expand=True)

        text_widget.insert(END, df.to_string())


    except Exception as e:
        messagebox.showerror('Error', str(e))


# Main window
window = Tk()
window.title('WESCRAP V(1.0)')
window.minsize(width=800, height=600)
window.maxsize(width=1920, height=1080)

# Widgets
url_label = Label(window, text="Enter URL Here:", font=("Calibri", 16), bg = 'teal', fg = 'white')
url_label.pack(pady=10)

url_entry = Entry(window, font=("Arial", 14), width=50, bg = 'teal', fg = 'light blue')
url_entry.pack()

tag_label = Label(window, text="Select Tag:", font=("Calibri", 16),bg = 'yellow', fg = 'red')
tag_label.pack(pady=10)

tag_var = IntVar()
tag_var.set(0)

tag_a = Radiobutton(window, text="Scrapping from 'a' tag", font=("Calibri", 14), variable=tag_var, value=0, bg = 'navy blue', fg = 'light green')
tag_a.pack()

tag_p = Radiobutton(window, text="Scrapping from 'p' tag", font=("Calibri", 14), variable=tag_var, value=1, bg = 'sky blue', fg = 'dark green')
tag_p.pack()

scrape_button = Button(window, text="Scrape Data", font=("Calibri", 16), command=scrape_data,bg = 'black', fg = 'light green')
scrape_button.pack(pady=10)

window.mainloop()
```
